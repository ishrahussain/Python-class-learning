```python
# as for alias
# with open(filename) as <variable>
with open('abc1.txt') as abc:
    content = abc.read()
    print(type(content))
    print(content)
```

```python
with open('abc1.txt') as abc:
    content = abc.readline() # Read until newline or EOF
    print(type(content))
    print(content)
```

```python
with open('abc1.txt') as abc:
    content = abc.readlines() # Return a list of lines from the stream
    print(type(content))
    print(content)
```

```python
with open('abc1.txt') as abc:
    content = abc.readlines()[0:3] # Slicing to get some lines
    print(type(content))
    print(content)
```

```python
with open('abc1.txt') as abc:
    content = abc.readlines()[0:3] # Slicing to get some lines
    abc.write("Pakistan Livelong") # cannot write as opened in read mode
    print(type(content))
    print(content)
```

```python
with open('a.txt', 'r+') as f:
    a = list(reversed(f.readlines()))
    print(a)
    b = list(reversed(['A\n', 'B\n', 'C\n']))
    print(b)
    c =  a + b
    print(c)
    d = "".join(c)
    f.seek(0)
    f.write(d)
```
